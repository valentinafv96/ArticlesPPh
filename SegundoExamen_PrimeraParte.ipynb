### La intención de esta primera parte del exámen consiste en generar un archivo **.json** con todos los artículos científicos de la Universidad de Antiquia que pertenezcan a mi grupo de investigación. Como dicho grupo es **GFIF** decidí que el mejor criterio para filtrarlos era buscar los artículos de la universidad que tuvieran asociado un nivel 1 de física de partículas.


---



```python
#@title <font color = 'green'>Librerías</font>

import json
import pandas as pd
from IPython.display import JSON

from google.colab import drive
drive.mount('/content/gdrive')
```

```python
#@title <font color = 'green'>Cargando archivo .json</font>

# Se carga el archivo .json con todos los artículos de Colombia suministrado por el profesor

file = open('gdrive/MyDrive/ComputacionAvanzada/openalexco.json','r')
co = json.load(file)
file.close()
```

```python
#@title <font color = 'green'>Filtrando datos</font>

# ARTÍCULOS DE LA UNIVERSIDAD DE ANTIOQUIA
# Se filtran los artículos con autores con alguna afiliación a esta universidad a través del ROR

co = [c for c in co if any(
    [any([i.get('ror')=='https://ror.org/03bp5hc83' for i in a.get('institutions')])
     for a in c.get('authorships')])
]

print(f'\nEl número de artículos de la Universidad de Antioquia en OpenAlex es {len(co)}\n')

# Artículos de Física de partículas
# Se filtran los artículos que en el primer nivel tengan asociada la palabra 'Particle Physics'

co = [d for d in co if len( {'Particle physics'}.intersection( 
       [ c.get('display_name') for c in d.get('concepts') if c.get('level')==1 ] 
          ) )>0 ]

print(f'El número de artículos en el área de Física de Partículas de la Universidad de Antioquia en OpenAlex es {len(co)}\n')
```

```python
#@title <font color = 'green'>Guardando datos</font>

# Se guardan los artículos filtrados pertenecientes a mi grupo de investigación en un archivo .json

f=open('gdrive/MyDrive/ComputacionAvanzada/openalexco_udea_pp.json','w')
json.dump(co,f)
f.close()
```

```python
#@title <font color = 'green'>Desplegrando los datos filtrados en en DataFrame</font>

f=open('gdrive/MyDrive/ComputacionAvanzada/openalexco_udea_pp.json','r')
pp=json.load(f)
f.close()

df = pd.DataFrame(pp)[['id','doi','publication_date','title']].sort_values('title').drop_duplicates(subset=['title']).reset_index(drop=True)

df
```
